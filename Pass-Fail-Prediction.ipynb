```python
import math
X=[1,2,3,4,5,6]
y=[0,0,0,1,1,1]
```

```python
m=len(X)
w=0
b=0
lr=0.1
iteration=1000
```

```python
def sigmoid(z):
    prd=1/(1+math.exp(-z))
```

```python
for _ in range(iteration):
    dw=0
    db=0
    cost=0
    for i in range(m):
        z=w*X[i]+b
        y_hat=sigmoid(z)
        cost+=-(y[i] * math.log(y_hat) + (1 - y[i]) * math.log(1 - y_hat))
        dw+=(y_hat-y[i])*X[i]
        db+=(y_hat-y[i])

        cost/=m
        w=w-lr*(dw/m)
        b=b-lr*(db/m)
        
```

```python
def pred(sol):
    z=w*sol+b
    y_hat=sigmoid(z)
    return 1 if y_hat >= 0.5 else 0
```

```python
sol=float(input("/n How many hours did study : "))
res=pred(sol)
if res==1:
    print("Pass")
else:
    print("Fail")
```
